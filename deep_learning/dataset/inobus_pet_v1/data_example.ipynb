```python
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt

import inobus_pet_v1

ds, info = tfds.load('inobus_pet_v1', with_info=True)
```

```python
get_label_name = info.features['label'].int2str
```

```python
train_ds, test_ds = ds['train'], ds['test']

plt.figure(figsize=(10, 10))
for i, sample in enumerate(train_ds.take(9)):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(sample['image'].numpy().astype('uint8'))
    plt.title(get_label_name(sample['label'].numpy()))
    plt.axis('off')
```

```python
plt.figure(figsize=(10, 10))
for i, sample in enumerate(test_ds.take(9)):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(sample['image'].numpy().astype('uint8'))
    plt.title(get_label_name(sample['label'].numpy()))
    plt.axis('off')
```
